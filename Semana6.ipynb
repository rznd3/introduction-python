# Exercício 1

Crie uma função em Python chamada `conta_par` que recebe como entrada uma lista de números inteiros e retorna a quantidade de números pares presentes nesta lista 

```python
def conta_par(lista):
  soma = 0
  for numero in lista:
    if numero%2 == 0:
      soma = soma +1

  return soma
```

```python
assert conta_par([]) == 0
assert conta_par([2,2,2,2,2]) == 5
assert conta_par([1,1,1,1,1]) == 0
assert conta_par([1,2,1,2,1,2]) == 3
```

# Exercício 2

Crie uma função em Python chamada `lista_par` que recebe como entrada uma lista de números inteiros e retorna uma nova lista contendo somente os números pares presentes na lista original. 

```python
def lista_par(lista):
  lista_pares = []
  for numero in lista:
    if numero%2 == 0:
      lista_pares.append(numero)

  return lista_pares
```

```python
assert lista_par([]) == []
assert lista_par([2,2,2,2,2]) == [2,2,2,2,2]
assert lista_par([1,1,1,1,1]) == []
assert lista_par([1,2,1,2,1,2]) == [2,2,2]
```

# Exercício 3

Crie uma função em python chamada `dobra_lista` que recebe como entrada $a$ uma lista de números e retorna uma nova $b$, em que cada elemento é o dobro do elemnto na mesma posição lista original (ou seja, $b[i]$  corresponde a $2*a[i]$)


```python
def dobra_lista(lista):
  lista_dobro = []
  for numero in lista:
    lista_dobro.append(numero*2)
  return lista_dobro
```

```python
assert dobra_lista([]) == []
assert dobra_lista([1,2,3]) == [2,4,6]
assert dobra_lista([10]) == [20]
assert dobra_lista([0,0,0]) == [0,0,0]
```

# Exercício 4

Crie uma função em python chamada `soma_lista` que recebe como entrada duas listas $a$ e $b$  de números de mesmo tamanho e retorna uma nova lista $c$, de tal maneira que $c[i]$ corresponde a  $a[i] + b[i]$ para todos os elmentos da lista

```python
def soma_lista(a,b):
  n = len(a)
  c = []
  for i in range(n):
    c.append(a[i]+b[i])

  return(c)
```

```python
assert soma_lista([],[]) == []
assert soma_lista([1,2,3],[3,2,1]) == [4,4,4]
assert soma_lista([1,1,1],[-1,-1,-1]) == [0,0,0]
assert soma_lista([0,0,0],[0,0,0]) == [0,0,0]
assert soma_lista([1],[2]) == [3]
```

# Exercício 5

Faça uma função em Python chamada `conta_igual` que recebe como entrada duas listas `a` e `b` de mesmo tamanho e conta quantos elementos dessas duas listas são iguais na mesma posição (ou seja, conta quantos elementos `a[i]` são iguais a `b[i]` para todos os elmentos dessas duas listas.

```python
def conta_igual(a,b):
  n = len(a)
  soma = 0
  for i in range(n):
    if a[i] == b[i]:
      soma = soma + 1

  return(soma)
```

```python
assert conta_igual([],[]) == 0
assert conta_igual([1,2,3],[3,2,1]) == 1
assert conta_igual([1,1,1],[-1,-1,-1]) == 0
assert conta_igual([0,0,0],[0,0,0]) == 3
assert conta_igual([1],[2]) == 0
```
