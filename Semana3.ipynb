# Exercício 1

Faça um programa que peça uma nota entre zero e dez. O seu programa deve então "validar" essa nota imprimindo uma mensagem de erro caso o valor esteje fora desse intervalo, e continue pedindo uma nova nota até o usuário informe um valor válido.

````python
nota = float(input('Digite uma nota de 0 a 10: '))

while nota < 0 or nota > 10:
  print('Valor invalido!')
  nota = float(input('Digite uma nota de 0 a 10: '))

print('Valor válido!')
````

# Exercício 2

Faça um programa que sorte um número inteiro entre zero e dez usando a biblioteca `random` (veja o método `randint`). O programa deve então solicitar que o usuário acerte o número, e ficar em um laço solicitando novos números até que o usuário acerte o número.

````python
import random

desconhecido = random.randint(0,10)

palpite = int(input("Digite um número de 0 a 10: "))

while desconhecido != palpite:
  print("Você errou, tente novamente: ")
  palpite = int(input('Digite um número de 0 a 10: '))

print("Você acertou!")
````

# Exercício 3

# Exercício 6

Faça um programa que leia o limite inferior e o superior de um intervalo, e conte quantos números detro deste intervalo são múltiplos, individualmente, de 3, 5, e de 3 e 5 ao mesmo tempo. 

````python
inferior = int(input("Digite o limite inferior: "))
superior = int(input("Digite o limite superior: "))

m_3_5 = 0
m_3 = 0
m_5 = 0

for i in range(inferior,superior):
  if i%15==0:
    m_3_5 += 1
  elif i%5==0:
    m_5 += 1
  elif i%3 ==0:
    m_3 += 1

print("Quantidade de múltiplos de 3 (exclusivamente):",m_3)
print("Quantidade de múltiplos de 5 (exclusivamente):",m_5)
print("Quantidade de múltiplos de 3 e 5 (simultâneo):",m_3_5)



````

# Exercício 7


Uma academia deseja fazer um senso entre os seus clientes para descobrir o mais alto, o mais baixo, o que tem o maior peso e o que tem o menor peso. Faça um programa que leia o nome, a altura e o peso de cada aluno. O programa deve parar de ler nomes quando o nome digitado for igual a vazio (""). O seu programa deve então imprimir o nome e o altura do aluno mais alto e mais baixo, e o nome e o peso do aluno de maior e de menor peso.

````python
nome = input("Digite o nome: ")
altura = float(input("Digite a altura: "))
peso = float(input("Digite o peso: "))

nome_mais_alto = nome
altura_mais_alto = altura

nome_mais_baixo = nome
altura_mais_baixo = altura

nome_maior_peso = nome
peso_maior = peso

nome_menor_peso = nome
peso_menor = peso


while nome!="":
  nome = input("Digite o nome: ")
  if nome == "":
    break
  altura = float(input("Digite a altura: "))
  peso = float(input("Digite o peso: "))

  if altura > altura_mais_alto:
    altura_mais_alto = altura
    nome_mais_alto = nome
  elif altura < altura_mais_baixo:
    altura_mais_baixo = altura
    nome_mais_baixo = nome

  if peso > peso_maior:
    peso_maior = peso
    nome_maior_peso = nome
  elif peso < peso_menor:
    peso_menor = peso
    nome_menor_peso = nome

print("Pessoa mais alta: ", nome_mais_alto,"com altura",altura_mais_alto)
print("Pessoa mais baixa: ", nome_mais_baixo,"com altura",altura_mais_baixo)
print("Pessoa mais pesada: ", nome_maior_peso,"com peso",peso_maior)
print("Pessoa menos pesada: ", nome_menor_peso,"com peso",peso_menor)
````
