# Exercício 1

Faça uma função chamada `reverte_palavras` que receba como entrada uma frase por meio de uma variável string e retorna uma nova frase com o reverso de cada palavra (cada palavra de trás pra frente). Considere que a frase não contém qualquer pontuação, somente palavras. 

```python
def reverte_palavras(frase):
  #quebra a frase em uma lista de palavras
  lista_palavras = frase.split()
  #cria uma lista para armazenar as palavras ao contrário
  lista_reversa = []
  # percorre a lista de palavras
  for palavra in lista_palavras:
    #adiciona a palavra ao contrario na lista de palavras reversas
    lista_reversa.append(palavra[::-1])
  #forma a frase novamente, com as palavras ao reverso
  return (" ".join(lista_reversa))
```

```python
assert reverte_palavras("o rato roeu a roupa do rei de roma")=="o otar ueor a apuor od ier ed amor"
assert reverte_palavras("adias a data da saída")=="saida a atad ad adías"
assert reverte_palavras("arara rara")=="arara arar"
```

# Exercício 2

Faça uma função chamada `tamanho_palavras` que receba como entrada uma frase por meio de uma variável string e retorna uma lista com o comprimento de cada palavra. Considere que a frase não contém qualquer pontuação, somente palavras. 

```python
def tamanho_palavras(frase):
  #quebra a frase em uma lista de palavras
  lista_palavras = frase.split()
  #lista para armazenar os tamanhos de palavras
  tamanhos_palavras = []
  #percorre a lista de palavras
  for palavra in lista_palavras:
    #adiciona o tamanho na lista de tamanhos
    tamanhos_palavras.append(len(palavra))
  #retorna a lista de tamanhos
  return tamanhos_palavras
```

```python
assert tamanho_palavras("o rato roeu a roupa do rei de roma")==[1, 4, 4, 1, 5, 2, 3, 2, 4]
assert tamanho_palavras("adias a data da saída")==[5, 1, 4, 2, 5]
assert tamanho_palavras("arara rara")==[5,4]
assert tamanho_palavras("constitucionalissimamente")==[25]
```

# Exercício 3

Faça uma função chamada `conta_palindromos` que recebe uma lista de palavras e conta quantos palíndromos existem nessa lista.

```python
def conta_palindromos(lista_palavras):
  # armazena o total de palindromos
  cont = 0
  #percorre a lsita de palavras
  for palavra in lista_palavras:
    # verifica se palavra é palindroma
    if palavra == palavra[::-1]:
      #se for verdade incrementa contador
      cont = cont+1
  #retorna a contagem
  return cont
```

```python
assert conta_palindromos(['sacas','saias','salas','socos','sopapos','sós'])==6
assert conta_palindromos(['oco','omissíssimo','osso','ovo'])==4
assert conta_palindromos(['o', 'rato', 'roeu', 'a', 'roupa', 'do', 'rei', 'de', 'roma'])==2
```

# Exercício 4

Crie uma função chamada `criptografa_par_impar` que é variação da Cifra de Cesar. Nessa versão, você deve aplicar a criptografia para cada palavra. Se o comprimento da palavra for par, você deve deslocar os caracteres em 3 posições. Se o comprimento for impar, você deve deslocar em 5 posições

```python
def criptografa_par_impar(frase):
  #quebra a frase em uma lista de palavras
  lista_palavras = frase.split()
  #cria uma lista para armazenar as palavras ao criptografadas
  lista_criptografada = []
  # percorre a lista de palavras
  for palavra in lista_palavras:
    # verifica se o tamanho da palavra é par ou impar
    if len(palavra)%2 == 0:
      # se for par, vamos delocar 3 posições
      deslocamento = 3
    else:
      # se for impar, 5 posições
      deslocamento = 5
    # string vazia para armazenar a palavra criptografada
    palavra_criptografa = ""
    # percorre a palavra original letra por letra
    for letra in palavra:
      # criptografa e adiciona a letra na palavra
      palavra_criptografa += chr(ord(letra)+ deslocamento)
    #adiciona a palavra criptografada na lista 
    lista_criptografada.append(palavra_criptografa)
  # forma a frase criptografada e retorna
  return(" ".join(lista_criptografada))
```

```python
assert criptografa_par_impar("o rato roeu a roupa do rei de roma")=="t udwr urhx f wtzuf gr wjn gh urpd"
assert criptografa_par_impar("adias a data da saída")=="finfx f gdwd gd xfòif"
assert criptografa_par_impar("arara rara")=="fwfwf udud"
```

# Exercício 5

Crie uma função chamada `decriptografa_par_impar` que faz a decriptografia do exercício anterior.

```python
def decriptografa_par_impar(frase):
  #quebra a frase em uma lista de palavras
  lista_palavras = frase.split()
  #cria uma lista para armazenar as palavras ao decriptogradas
  lista_decriptografada = []
  # percorre a lista de palavras
  for palavra in lista_palavras:
    # verifica se o tamanho da palavra é par ou impar
    if len(palavra)%2 == 0:
      # se for par, vamos delocar 3 posições
      deslocamento = 3
    else:
      # se for impar, 5 posições
      deslocamento = 5
    # string vazia para armazenar a palavra criptografada
    palavra_decriptografa = ""
    # percorre a palavra original letra por letra
    for letra in palavra:
      # decriptografa e adiciona a letra na palavra
      palavra_decriptografa += chr(ord(letra) - deslocamento)
    #adiciona a palavra criptografada na lista 
    lista_decriptografada.append(palavra_decriptografa)
  # forma a frase criptografada e retorna
  return(" ".join(lista_decriptografada))
```

```python
assert decriptografa_par_impar("t udwr urhx f wtzuf gr wjn gh urpd")=="o rato roeu a roupa do rei de roma"
assert decriptografa_par_impar("finfx f gdwd gd xfòif")=="adias a data da saída"
assert decriptografa_par_impar("fwfwf udud")=="arara rara"
```
