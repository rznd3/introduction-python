# Exercício 1:

Em matemática, a fórmula de Leibniz para $\pi$, que leva o nome de Gottfried Wilhelm Leibniz, estabelece que
$$ \pi \approx 4\left(1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \frac{1}{9} - \cdots \right)$$
ou, em notação matemática:
$$ \pi \approx 4  \sum_{i=0}^n \frac{(-1)^i}{2i+1} $$
Essa série converge para $\pi$ para $n\rightarrow\infty$. 

Crie uma função chamada `pi` que recebe como entrada um valor de $n$ e, usando um laço, calcula e retorna o valor aproximado de $\pi$ usando a fórmula de Leibniz com $i \in 0..n$. A célula com o comando assert será usada para testar a sua implementação

```python
def pi(n):
  soma = 0
  for i in range(n+1):
    soma = soma + ((-1)**i)/(2*i + 1)

  return 4*soma
```

```python
from math import isclose

assert isclose(pi(3),2.8952380952380956,abs_tol=1e-8)
assert isclose(pi(10),3.232315809405594,abs_tol=1e-8)
assert isclose(pi(100),3.1514934010709914,abs_tol=1e-8)
assert isclose(pi(1000),3.1425916543395442,abs_tol=1e-8)
```

# Exercício 2

Dizemos que um número natural é triangular se ele e produto de três número naturais consecutivos. Por exemplo: 120 é triangular, pois $4*5*6 = 120$.  2730 é triangular,
pois $13*14*15 = 2730$. Dado um inteiro  não-negativo `n`, crie uma função chamada `triangular`  para verificar se `n` é triangular ou não. A função deve-se devolver `True` se o número for triangular, e `False` caso contrário.

```python
def triangular(n):
  for i in range(n//3):
    if i * (i+1) * (i+2) == n:
      return True
  return False
```

```python
assert triangular(120) == True
assert triangular(2730) == True
assert triangular(50) == False
assert triangular(1500) == False
```

# Exercício 3

O jogo pedra-papel-tesoura (também chamado de jokenpô) é um jogo de mãos recreativo e simples para duas ou mais pessoas. O jogo é frequentemente empregado como método de seleção (como na escolha de equipes para a prática desportiva, por exemplo). 

No pedra-papel-tesoura, os jogadores devem simultaneamente esticar a mão, na qual cada um formou um símbolo (que significa pedra, papel ou tesoura). Então, os jogadores comparam os símbolos para decidir quem ganhou, da seguinte forma:

- Pedra ganha da tesoura.
- Tesoura ganha do papel.
- Papel ganha da pedra.

A pedra é simbolizada por um punho fechado; a tesoura, por dois dedos esticados; e o papel, pela mão aberta. Caso dois jogadores façam o mesmo gesto, ocorre um empate. 

Faça uma função chamada `jokenpo` que recebe como entrada a jagada de dois jogadores (representado pelas strings `pedra`, `papel` ou `tesoura`) e retorna 1 se o primeiro jogador ganhou, 2 se o segundo jogador ganhou, e 0 em caso de empate.

```python
def jokenpo(jogador1, jogador2):
  if (jogador1 == 'pedra' and jogador2 == 'tesoura') or (jogador1 == 'tesoura' and jogador2 == 'papel') or (jogador1 == 'papel' and jogador2 == 'pedra'):
    return 1
  elif (jogador2 == 'pedra' and jogador1 == 'tesoura') or (jogador2 == 'tesoura' and jogador1 == 'papel') or (jogador2 == 'papel' and jogador1 == 'pedra'):
    return 2
  else:
    return 0
```

```python
assert jokenpo('pedra','tesoura')==1
assert jokenpo('tesoura','pedra')==2
assert jokenpo('tesoura','papel')==1
assert jokenpo('papel','tesoura')==2
assert jokenpo('papel','pedra')==1
assert jokenpo('pedra','papel')==2
assert jokenpo('pedra','pedra')==0
assert jokenpo('tesoura','tesoura')==0
assert jokenpo('papel','papel')==0
```

# Exercício 4

Faça um função chamada `conta_vogal` que, usando laços, conta quantas vogais existem em uma palavra. 

Dica: seja `s` uma variável que contém uma letra. Você pode testar se essa variável é uma vogal com a expressão `s in ['a','e','i','o','u']`. 

```python
def conta_vogal(palavra):
  soma = 0
  for letra in palavra:
    if letra in ['a','e','i','o','u']:
      soma = soma + 1
  
  return soma
```

```python
assert conta_vogal('abacate') == 4
assert conta_vogal('chuchu') == 2
assert conta_vogal('vogal') == 2
assert conta_vogal('paralelepipedo') == 7
```

# Exercício 5

A [fórmula de Herão](https://pt.wikipedia.org/wiki/Teorema_de_Her%C3%A3o) para o cálculo da área de um triângulo é dada por
$$A = \sqrt{p(p-a)(p-b)(p-c)}$$
em que $a$, $b$ e $c$ são os lados de um triângulo, e p é o semi-perímetro, dado pela fórmula:
$$p=\frac{a+b+c}{2}$$
Crie uma função chamada `area_triangulo` que receba os três lados de um triângulo e, usando a fórmula de Herão, calcule e retorne a área desse triângulo.


```python
from math import sqrt
def area_triangulo(a,b,c):
  p = (a+b+c)/2
  A = sqrt(p*(p-a)*(p-b)*(p-c))
  return(A)
```

```python
from math import isclose

assert isclose(area_triangulo(4,5,6),9.921567416492215,abs_tol=1e-8)
assert isclose(area_triangulo(8,9,10),34.197039345533994,abs_tol=1e-8)
assert isclose(area_triangulo(1,2,3),0,abs_tol=1e-8)
assert isclose(area_triangulo(10,20,30),0,abs_tol=1e-8)
```
